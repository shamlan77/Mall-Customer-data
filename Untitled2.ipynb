```python
# prompt: Identify and handle missing values using .isnull() in Python

# Identify missing values
missing_values = df.isnull().sum()
print("Missing Values per column:\n", missing_values)

# Handle missing values (example: fill with mean for numerical columns)
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

# Alternative handling (example: drop rows with missing values)
# df.dropna(inplace=True)

# Verify the handling
missing_values_after = df.isnull().sum()
print("\nMissing Values after handling:\n", missing_values_after)
```
